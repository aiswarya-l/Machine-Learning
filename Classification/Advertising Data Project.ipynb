In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

### Split the data into features and target variables (X and y)

```python
df.head(2)
```

```python
df = df.drop(columns =['Ad Topic Line','City','Country','Timestamp'])
```

```python
df.head()
```

```python
X = df.drop(columns='Clicked on Ad')
```

```python
y = df['Clicked on Ad']
```

### Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X
```

```python
xcolumns = X.columns
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
X = pd.DataFrame(X,columns = xcolumns)
```

```python
X.head()
```

```python
X.drop(columns=['Month','Year','Date','Hour'],inplace = True)
```

```python
X.head()
```

### Split the data into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

### Create a Logistic Regression model and train it

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

### Check the accuracy of our model

```python
model.score(X_train,y_train)
```

### Make prediction using the X_test

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

### Check how accurate the prediction is:

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_pred)
```

### Check the confusion matrix

```python
metrics.confusion_matrix(y_test,y_pred)
```

### Plot the confusion matrix on a heatmap

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_pred),annot = True, fmt= 'd')
```

### Create a classification report for the model

```python
print(metrics.classification_report(y_test,y_pred))
```
